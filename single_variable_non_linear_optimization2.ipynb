```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
  return x*4 + 3*x*2 + 10
def df(x):
  return 4*x**3 + 6*x
learning_rate = 0.01
iterations = 2
x = 2.0
history_x = [x]
history_y = [f(x)]
print("Manual Calculation:")
print(f"Iteration 0: x = {x:.4f}, f(x) = {f(x):.4f}")
for i in range(iterations):
  grad = df(x)
  x = x - learning_rate * grad
  history_x.append(x)
  history_y.append(f(x))
  print(f"Iteration {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}")
print("\nResult after Gradient Descent:")
print(f"Optimal x: {x:.4f}")
print(f"Minimum value: {f(x):.4f}")
x_plot = np.linspace(-2, 2, 100)
y_plot = f(x_plot)
plt.plot(x_plot, y_plot, label='f(x)')
plt.scatter(history_x, history_y, color='red', label='Steps')
plt.scatter(history_x[-1], history_y[-1], color='green', marker='*', s=100, label='Minimum')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent')
plt.legend()
plt.grid(True)
plt.show()
```
