```python
import numpy as np
def f(x, y):
    return 3*x**2 + 5*np.exp(-y) + 10
def gradients(x, y):
    df_dx = 6 * x
    df_dy = -5 * np.exp(-y)
    return df_dx, df_dy
def gradient_descent(lr=0.1, iterations=100):
    x, y = 1.0, 1.0  # initial guess
    for i in range(iterations):
        grad_x, grad_y = gradients(x, y)
        x -= lr * grad_x
        y -= lr * grad_y
        print(f"Iteration {i+1}: x = {x:.5f}, y = {y:.5f}, f(x, y) = {f(x, y):.5f}")
    return x, y, f(x, y)
min_x, min_y, min_val = gradient_descent()
print(f"\nMinimum at x = {min_x:.5f}, y = {min_y:.5f} with value = {min_val:.5f}")
```
