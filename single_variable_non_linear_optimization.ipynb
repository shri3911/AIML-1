```python
import numpy as np
def f(x):
  return x**2 + 5
def df(x):
  return 2*x
def gradient_descent(initial_x, learning_rate, tolerance, max_iterations):
  x = initial_x
  x_history = [x]
  fx_history = [f(x)]
  for i in range(max_iterations):
    gradient = df(x)
    x_new = x - learning_rate * gradient
    x_history.append(x_new)
    fx_history.append(f(x_new))
    if np.abs(x_new - x) < tolerance:
      break
    x = x_new
  return x, x_history, fx_history
initial_x = 2.0
learning_rate = 0.01
tolerance = 1e-6
max_iterations = 100
optimal_x, x_history, fx_history = gradient_descent(initial_x, learning_rate, tolerance, max_iterations)
print(f"Optimal x found: {optimal_x:.4f}")
print(f"Minimum value of f(x): {f(optimal_x):.4f}")
import matplotlib.pyplot as plt
plt.plot(x_history, fx_history, marker='o')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gradient Descent Optimization")
plt.grid(True)
plt.show()
```
